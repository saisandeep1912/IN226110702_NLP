# Task 2:Build Advanced Preprocessing Function


```python
import re
def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""
    # Lowercase
    text = text.lower()
    # Remove URLs & emails
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    # Remove numbers
    text = re.sub(r"\d+", "", text)
    # Handle repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)
    # Remove special characters (keep words)
    text = re.sub(r"[^a-z\s]", "", text)
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    tokens = text.split()
    # Remove short tokens (≤2) except "no", "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]
    cleaned_sentence = " ".join(tokens)
    return tokens, cleaned_sentence
```

# 3.Task 3: Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in test_sentences:
    tokens, clean = preprocess_text(text)
    print("Original:", text)
    print("Tokens:", tokens)
    print("Clean:", clean)
    print("-"*50)
```

# Task 4: Token Analytics


```python
def token_analytics(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total

    return total, unique, round(avg_len, 2)
```

```python
for text in test_sentences:
    tokens, clean = preprocess_text(text)
    total, unique, avg = token_analytics(tokens)

    print(f"Text: {text}")
    print(f"Total Tokens: {total}, Unique: {unique}, Avg Length: {avg}")
```

# Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for text in test_sentences:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent words:", counter.most_common(10))

print("Least frequent words:",
      sorted(counter.items(), key=lambda x: x[1])[:5])
```

# Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
output = full_pipeline(test_sentences)
print(output)
```

# Task 7: Error Handling


```python
import re

def preprocess_text(text):
    try:
        if not isinstance(text, str):
            return [], ""
        text = text.strip()
        if text == "":
            return [], ""
        text = text.lower()
        text = re.sub(r"http\S+|www\S+|https\S+", "", text)
        text = re.sub(r"\S+@\S+", "", text)
        text = re.sub(r"\d+", "", text)
        text = re.sub(r"(.)\1{2,}", r"\1", text)
        text = re.sub(r"[^a-z\s]", "", text)
        text = re.sub(r"\s+", " ", text).strip()
        tokens = text.split()
        tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]
        cleaned_sentence = " ".join(tokens)
        return tokens, cleaned_sentence
    except Exception as e:
        print(f"Error processing text: {text} | Error: {e}")
        return [], ""
```

```python
edge_cases = [
    "",                     # empty string
    "     ",                # only spaces
    "123456789",            # only numbers
    "😍😍🔥🔥",              # only emojis
    None,                  # None input
    "@@@###$$$",           # only symbols
    "NO!!!",               # important keyword
    "not good",            # negation case
]

for case in edge_cases:
    tokens, clean = preprocess_text(case)
    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-" * 50)
```
